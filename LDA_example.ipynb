```python
import gensim
from gensim.models import CoherenceModel
import matplotlib.pyplot as plt
```

```python
import pandas as pd
from konlpy.tag import Okt

# 데이터 로드
okt = Okt()
data = data.dropna(subset=['content'])  # 'content' 열에서 NaN 값이 있는 행을 제거

# 데이터 전처리: 'content' 열을 가정합니다. 실제 데이터의 컬럼명을 확인해주세요.
data['nouns'] = data['content'].apply(lambda text: [word for word, tag in okt.pos(text) if tag in ['Noun']])

# 한국어 스톱워드 리스트
stopwords = pd.read_csv("https://raw.githubusercontent.com/yoonkt200/FastCampusDataset/master/korean_stopwords.txt", header=None).values.flatten().tolist()
additional_stopwords = ['하다','있다','으로','되다','이다','대다',')','만에','않다','만에','하고','가장','이후','통해','대비','기록','가다', '의무','위해',
                        '구역','단지','전체','없다','올해','일반','적용','지수','마련','기준','설치','비중','에는','정보','시스템','지역','부터','건수',
                        '에서는', '이하','규모','그룹','2021년','최근','관련','주차','기자','지다','입자','때문','가율','상품','이행','이상','사실','발표',
                        '에게','개월','사진','대상','예정','한국','조성','업체','시설','제공','센터','경우','가구','사업','지원','건물']
stopwords.extend(additional_stopwords)

# 텍스트 전처리: 스톱워드 제거 및 길이가 1 이하인 단어 제거
data['filtered_nouns'] = data['nouns'].apply(lambda tokens: [token for token in tokens if token not in stopwords and len(token) > 1])

# 사전과 코퍼스 생성
from gensim import corpora, models
dictionary = corpora.Dictionary(data['filtered_nouns'])
dictionary.filter_extremes(no_below=5, no_above=0.5)

corpus = [dictionary.doc2bow(text) for text in data['filtered_nouns']]

# LDA 모델 학습
ldamodel = models.ldamodel.LdaModel(corpus, num_topics=12, id2word=dictionary, passes=15)


# 각 토픽의 단어들 출력
topics = ldamodel.print_topics(num_words=20)
for topic in topics:
    print(topic)
```

```python
def compute_coherence_values(dictionary, corpus, texts, start, limit, step):
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        print(f"Training LDA model with {num_topics} topics...")
        model = gensim.models.ldamodel.LdaModel(corpus=corpus, num_topics=num_topics, id2word=dictionary, passes=15)
        model_list.append(model)
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())

    return model_list, coherence_values

# 한국어 명사만 추출된 텍스트 데이터를 리스트로 변환
texts = [doc for doc in data['filtered_nouns']]

# Coherence 값을 계산
start, limit, step = 2, 15, 1
model_list, coherence_values = compute_coherence_values(dictionary=dictionary, corpus=corpus, texts=texts, start=start, limit=limit, step=step)

# 최적의 토픽 수를 찾기 위한 응집성 점수 시각화
x = range(start, limit, step)
plt.plot(x, coherence_values)
plt.xlabel("Number of Topics")
plt.ylabel("Coherence Score")
plt.title("Coherence Score by Number of Topics")
plt.show()

# 최적의 토픽 수 출력
optimal_index = coherence_values.index(max(coherence_values))
optimal_model = model_list[optimal_index]
print(f"Optimal number of topics: {optimal_model.num_topics}")
```

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from gensim.models import LdaModel
from sklearn.metrics.pairwise import cosine_similarity




# 토픽의 단어 분포를 벡터 형태로 추출
topic_terms = ldamodel.get_topics()  # Shape: (num_topics, vocabulary_size)

# 토픽 간 코사인 유사도 계산
cosine_sim = cosine_similarity(topic_terms)

# 네트워크 생성
G = nx.Graph()

# 노드 추가 (토픽을 노드로 추가)
for i in range(len(topic_terms)):
    G.add_node(i, label=f"Topic {i+1}")

# 에지 추가 (유사도가 특정 임계값 이상인 토픽 간에 에지를 추가)
threshold = 0.4 # 유사도 임계값 설정
for i in range(len(cosine_sim)):
    for j in range(i + 1, len(cosine_sim)):
        if cosine_sim[i][j] > threshold:
            G.add_edge(i, j, weight=cosine_sim[i][j])

# 네트워크 시각화
pos = nx.spring_layout(G)  # 노드의 위치 결정
labels = nx.get_node_attributes(G, 'label')
weights = nx.get_edge_attributes(G, 'weight')

nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=700)
nx.draw_networkx_edges(G, pos, width=[w * 10 for w in weights.values()])
nx.draw_networkx_labels(G, pos, labels, font_size=12)
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): f"{w:.2f}" for (u, v), w in weights.items()})
plt.title('Topic Similarity Network')
plt.axis('off')
plt.show()
```

```python
import pandas as pd

# 토픽 모델과 코퍼스를 가정
# ldamodel = ... (LDA 모델 인스턴스)
# corpus = ... (문서 코퍼스)

# 문서별 토픽 비중을 저장할 빈 데이터프레임 생성
# 토픽 수에 따라 데이터프레임의 열을 조정해야 할 수 있습니다.
num_topics = 12
topic_columns = [f'Topic {i}' for i in range(num_topics)]
df1 = pd.DataFrame(columns=topic_columns)

# 각 문서의 토픽 분포 추출
rows = []  # 데이터프레임에 추가할 행을 저장할 리스트

for i, row in enumerate(ldamodel[corpus]):
    # 각 토픽의 초기 확률을 0으로 설정
    topic_probs = {f'Topic {topic}': 0 for topic in range(num_topics)}
    
    # 각 토픽에 대한 확률 업데이트
    for topic, prob in row:
        topic_probs[f'Topic {topic}'] = prob
    
    # rows 리스트에 토픽 확률 딕셔너리 추가
    rows.append(topic_probs)

# 리스트를 이용해 데이터프레임 생성
df1 = pd.DataFrame(rows)

# 데이터프레임 출력
print(df1)
```
